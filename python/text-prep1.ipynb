We can pull out all tokens that spaCy identified, and create a list that contains only those words that are considered "Out of Vocabulary". These are not precisely spelling errors, but this  

```python
tmp = correggio_doc[4]
print("Total tokens: " + str(len(correggio_doc)))

# Preview what the is_oov attribute looks like: 
print('\n' + "Preview of first 15 words:")
for j in range(15):
    print(correggio_doc[j].text,correggio_doc[j].is_oov)
```

```python
# Build a list of OOV words 
oov_words = list()
for j in range(len(correggio_doc)):
    if correggio_doc[j].is_oov is True:
        oov_words.append(correggio_doc[j].text)
        # print(correggio_doc[j].text,correggio_doc[j].is_oov)

# Distill the unique words:
oov_words_unique = list(set(oov_words))
print("Total unique OOV words: " + str(len(oov_words_unique)))
list(oov_words_unique)
```
